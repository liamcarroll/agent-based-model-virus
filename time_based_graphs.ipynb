```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
```

# Simulation 1

```python
deaths_df1 = pd.read_csv('./data/sim1/deaths.csv', header=0)
cases_df1 = pd.read_csv('./data/sim1/cases.csv', header=0)
immune_df1 = pd.read_csv('./data/sim1/immune.csv', header=0)
```

```python
plt.plot(deaths_df1['deaths'])
plt.savefig('./pics/sim1/deaths.png', bbox_inches='tight')
```

```python
plt.plot(cases_df1['cases'])
plt.savefig('./pics/sim1/cases.png', bbox_inches='tight')
```

```python
plt.plot(immune_df1['immune'])
plt.savefig('./pics/sim1/immune.png', bbox_inches='tight')
```
